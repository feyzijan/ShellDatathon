Plot sales across time

```python
df_train = df.drop(labels="Customer",axis=1)
```

Preprocess a little

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(handle_unknown='ignore')
```

```python
data = df_train.copy()

from sklearn.preprocessing import LabelBinarizer

jobs_encoder = LabelBinarizer()
jobs_encoder.fit(data['Area'])
transformed = jobs_encoder.transform(data['Area'])
ohe_df = pd.DataFrame(transformed)
data = pd.concat([data, ohe_df], axis=1).drop(["Area"], axis=1)
```

```python
data["Petrol"] =  (data["Product"] == "Benzin").astype(int)
```

```python
data.drop("Product",axis=1,inplace=True)
```

```python
data.drop("WeekStarting",axis=1,inplace=True)
```

### Fit svm

```python
from sklearn.svm import SVR
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
```

```python
regr = make_pipeline(StandardScaler(), SVR(C=1.0, epsilon=0.2))
```

```python
label = data["Litres"]
data.drop("Litres",axis=1,inplace=True)
```

```python
data.drop("Season",axis=1,inplace=True)
data.drop("WeekNumber",axis=1,inplace=True)
data.dtypes
```

```python
regr.fit(data, label)
```
