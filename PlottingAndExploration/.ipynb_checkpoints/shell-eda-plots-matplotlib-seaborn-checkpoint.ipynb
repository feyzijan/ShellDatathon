```python
def grab_col_names(dataframe, cat_th=10,  car_th=20, report=False):
    # cat_cols, cat_but_car
    cat_cols = [col for col in dataframe.columns if str(dataframe[col].dtypes) in ["category", "object", "bool"]]

    num_but_cat = [col for col in dataframe.columns if dataframe[col].nunique() < cat_th and dataframe[col].dtypes in ["int", "float"]]

    cat_but_car = [col for col in dataframe.columns if
                   dataframe[col].nunique() > car_th and str(dataframe[col].dtypes) in ["category", "object"]]

    # final categorical variables
    cat_cols = cat_cols + num_but_cat
    cat_cols = [col for col in cat_cols if col not in cat_but_car]

    # final numerical variables
    num_cols = [col for col in dataframe.columns if dataframe[col].dtypes in ["int", "float"]]
    num_cols = [col for col in num_cols if col not in cat_cols]

    # report
    if report:
        print(f"Observations: {dataframe.shape[0]}")
        print(f"Variables: {dataframe.shape[1]}")
        print(f'cat_cols: {len(cat_cols)}')
        print(f'num_cols: {len(num_cols)}')
        print(f'cat_but_car: {len(cat_but_car)}')
        print(f'num_but_cat: {len(num_but_cat)}')

    return cat_cols, num_cols, cat_but_car


dfs = [train, test, price]
for d in dfs:
    name_of_df = " "+ get_df_name(d).upper() + " "
    print(f" {colored(name_of_df.center(50, '~'), 'yellow', attrs=['reverse'])}", "\n")
    cat_cols, num_cols, cat_but_car = grab_col_names(d, report=True)
    print(f"\n{colored('Numerical Columns:','blue', attrs=['reverse'])} {num_cols}\n\n{colored('Categorical Columns:','magenta', attrs=['reverse'])} {cat_cols}\n\n"
          f"{colored('Cardinal Columns:','cyan', attrs=['reverse'])}{cat_but_car}\n")
```

# 2. Kategorik Değişken Analizi 🧐

```python
def cat_summary(dataframe, col_name, plot=False):
    print(pd.DataFrame({col_name: dataframe[col_name].value_counts(),
                        "Ratio": 100 * dataframe[col_name].value_counts() / len(dataframe)}))
    print("##########################################")

    if plot:
        explode = []
        for i in range(len(dataframe[col_name].value_counts())):
            explode.append(i*0.05)
        f,ax=plt.subplots(1,2, figsize=(10,4))
        dataframe[col_name].value_counts().plot.pie(explode=explode,autopct='%1.1f%%',ax=ax[0],shadow=True)
        ax[0].set_title(f"Distribution of {col_name}", fontdict={"fontsize":15})
        ax[0].set_ylabel("")
        sns.countplot(x=dataframe[col_name], data=dataframe, ax=ax[1])
        ax[1].set_title(f"Count of {col_name}", fontdict={"fontsize":15})
        ax[1].set_xlabel("")
        plt.tight_layout()
        plt.show()
```

```python
for d in dfs:
    cat_cols, num_cols, cat_but_car = grab_col_names(d, report=False)
    for col in cat_cols:
        cat_summary(d, col, plot=True)
```

# 4. Hedef Değişken Analizi 🎯

```python
def target_summary_with_cat(dataframe, target, categorical_col, plot=False):
    df = pd.DataFrame({"TARGET_MEAN": dataframe.groupby(categorical_col)[target].mean()})
    display(df)
    
    if plot:
        colors = sns.color_palette('pastel')[0:5]
        plt.title(f"{categorical_col} Rates")
        plt.pie(x=df.reset_index()["TARGET_MEAN"],
                labels = df.reset_index()[categorical_col],
                data=df,
                colors = colors, autopct='%.0f%%')
        plt.show()
        
    print("".center(50,"~"))



print(f" {colored(' TRAIN '.center(70, '~'), 'red', attrs=['reverse'])}", "\n")
cat_cols, num_cols, cat_but_car = grab_col_names(train, report=False)
cat_cols.append("City Code")
for col in cat_cols:
    target_summary_with_cat(train, "Litres", col, plot=True)
```

# 5. Rare Analysis 💎

```python
def rare_analyser(dataframe, target, cat_cols):
    for col in cat_cols:
        print(col, ":", len(dataframe[col].value_counts()))
        print(pd.DataFrame({"COUNT": dataframe[col].value_counts(),
                            "RATIO": dataframe[col].value_counts() / len(dataframe),
                            "TARGET_MEAN": dataframe.groupby(col)[target].mean()}).sort_values("RATIO"), end="\n\n\n")

print(f" {colored(' TRAIN '.center(70, '~'), 'red', attrs=['reverse'])}", "\n")
cat_cols, num_cols, cat_but_car = grab_col_names(train, report=False)
cat_cols.append("City Code")
for col in cat_cols:
    rare_analyser(train, "Litres", cat_cols)
```
