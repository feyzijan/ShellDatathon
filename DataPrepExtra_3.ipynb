Sort data properly

```python
df.sort_values(by=["Product","Customer","WeekStarting"],ascending=[False,True,True],inplace=True)
df.columns.shape
```

Reorder columns for later ease

```python
df = df[['Customer', 'WeekStarting','Litres',
       'WeekendHolidays', 'WeekdayHolidays', 'Area', 'Product', 'Season',
       'WeekNumber', 'Price(t-1)','Price%Change',
       'NewCars', 'GDPGrowth%','MoMInflation%', 'WoWUSDChange%','Litres(w-51)',  'Litres(w-52)',
       'Litres(w-53)', 'Litres(w-54)', 'Litres(w-51_average_city)','Litres(w-52_average_city)',
       'Litres(w-53_average_city)', 'Litres(w-54_average_city)',
       'Litres(w-51_average_area)','Litres(w-52_average_area)', 'Litres(w-53_average_area)',
       'Litres(w-54_average_area)']]
```

Standard Scaling

```python
from sklearn.preprocessing import StandardScaler

X = df.drop(labels=['Customer', 'WeekStarting','Litres',
       'WeekendHolidays', 'WeekdayHolidays', 'Area', 'Product', 'Season'],axis=1)
X_ids = df[["Customer","WeekStarting"]]
y = df["Litres"]

scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
```

Assign scaled data to df

```python
X_scaled.shape
```

```python
cols = df.columns
df[cols[8:]] = X_scaled
df.head()
```

One Hot encode stuff

```python
from sklearn.preprocessing import OneHotEncoder

OneHotEncoder(
    categories='auto',  # Categories per feature
    drop=None, # Whether to drop one of the features
    sparse=True, # Will return sparse matrix if set True
    handle_unknown='error' # Whether to raise an error 
)    
```

```python
ohe = OneHotEncoder()
transformed = ohe.fit_transform(df[["Product","Area","Season"]])
```

```python
transformed.toarray()
```

```python
ohe.categories_
```

```python
df[ohe.categories_[0]] = transformed.toarray()[:,:2]
df[ohe.categories_[1]] = transformed.toarray()[:,2:9]
df[ohe.categories_[2]] = transformed.toarray()[:,9:]
```

```python
# Drop encoded columns
df.drop(labels=["Area","Product","Season"],axis=1,inplace=True)
```

```python
df.head()
```

Drop First Year

```python
df = df[df["WeekStarting"] > "2015-01-01"]
df = df.reset_index(drop=True)
df.head()
```

Separate TrainingAndTestData

```python
df_training = df[df.WeekStarting < '2018-12-31'].reset_index()
df_test = df[df.WeekStarting > '2018-12-31'].reset_index()
```

Save data

```python
df_training.to_pickle('TrainingDataFinalized.pkl')
df_test.to_pickle('TestDataFinalized.pkl')
```
