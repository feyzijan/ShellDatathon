New car Registration Data

```python
df_cars = pd.read_excel("ExternalData\TUIKKaraTasıtYeniKayit.xls")
```

```python
df_cars
```

- Combine date columns

```python
df_cars["Date"] = df_cars["Year"].astype(str) + "-" + df_cars['Month'].astype(str).str[0:2] + "-01"
df_cars.drop(labels=["Month","Year"],axis=1,inplace=True)
```

```python
df_cars.head()
```

```python
df_cars.columns
```

Combine into following regions
- Marmara Bölgesi    
    - TR2,TR4,TR1 
- Ege Bölgesi  
    - TR3            
- İç Anadolu Bölgesi 
    - TR5,TR7       
- Akdeniz Bölgesi
    - TR6               
- Karadeniz Bölgesi    
    - TR8,TR9        
- Güneydoğu Anadolu Bölgesi  
    - TRC   
- Doğu Anadolu Bölgesi    
    - TRA,TRB      

```python
# Create new columns with the sum of each region
df_cars["Marmara Bölgesi"] = df_cars[['Batı Marmara-TR2','Doğu Marmara-TR4','İstanbul-TR1']].sum(axis=1)
df_cars["Ege Bölgesi"] = df_cars['Ege-TR3']
df_cars["İç Anadolu Bölgesi"] = df_cars[['Batı Anadolu-TR5', 'Orta Anadolu-TR7']].sum(axis=1)
df_cars["Akdeniz Bölgesi"] = df_cars['Akdeniz-TR6']
df_cars["Karadeniz Bölgesi"] = df_cars[['Doğu Karadeniz-TR9','Batı Karadeniz-TR8']].sum(axis=1)
df_cars["Güneydoğu Anadolu Bölgesi"] =  df_cars['Güneydoğu Anadolu-TRC']
df_cars["Doğu Anadolu Bölgesi"] = df_cars[['Kuzeydoğu Anadolu-TRA','Ortadoğu Anadolu-TRB']].sum(axis=1)
```

```python
# drop old columns
df_cars = df_cars[["Date","Akdeniz Bölgesi", "İç Anadolu Bölgesi","Karadeniz Bölgesi","Marmara Bölgesi", "Ege Bölgesi", "Güneydoğu Anadolu Bölgesi",
"Doğu Anadolu Bölgesi"]]
```

```python
df_cars
```

```python
# Convert to datetime object and add new column indicating when the data is available (t+2 months)
df_cars["Date"]= pd.to_datetime(df_cars['Date'])
df_cars["DataReleaseDate"] = df_cars["Date"] + pd.to_timedelta(50,unit='d')
```

```python
df_cars.head()
```

```python
# Set the data release date to nearest sunday
df_cars["DataReleaseDate"] = df_cars["DataReleaseDate"] - pd.to_timedelta(df_cars["DataReleaseDate"].dt.dayofweek, unit='d') +  pd.to_timedelta(6, unit='d')
```

```python
# pivot table so its easier to merge
df_cars = df_cars.melt(id_vars=["Date", "DataReleaseDate"],
var_name="Region",
value_name="Cars")
```

```python
df_cars = df_cars[df_cars["DataReleaseDate"] > "2013-12-01" ]
df_cars.sort_values(by=["Region","Date"],inplace=True)
df_cars.reset_index(inplace=True,drop=True)
```

```python
df_cars
```

- Put these into a weekly df

```python
df_dates = pd.date_range('2014-01-05', '2019-12-30', freq='1W').to_frame(name='Date').reset_index(drop=True)
df_dates = pd.DataFrame(np.repeat(df_dates.values, 7, axis=0))
df_dates.columns = ["Date"]
```

```python
df_dates["Region"] = ["Akdeniz Bölgesi", "İç Anadolu Bölgesi","Karadeniz Bölgesi","Marmara Bölgesi", "Ege Bölgesi", "Güneydoğu Anadolu Bölgesi",
"Doğu Anadolu Bölgesi"]*313
```

```python
df_dates.sort_values(by=["Region","Date"],inplace=True)
```

```python
df_dates.reset_index(inplace=True,drop=True)
```

```python
df_dates["Cars"] = 0

for i in range(len(df_dates)):
    for k in range(len(df_cars)-1):
        if df_dates.loc[i,"Region"] == df_cars.loc[k,"Region"]:
            if df_dates.loc[i,"Date"] > df_cars.loc[k,"DataReleaseDate"] and df_dates.loc[i,"Date"] <= df_cars.loc[k+1,"DataReleaseDate"]:
                #print(df_dates.loc[i,"Date"], 
                #df_cars.loc[k,"DataReleaseDate"],
                #df_cars.loc[k+1,"DataReleaseDate"])
                df_dates.loc[i,"Cars"] = df_cars.loc[k,"Cars"]
```

```python
# Add one day so it matches with the training/test data
df_dates["Date"] = df_dates["Date"] + pd.to_timedelta(1,unit='d')
```

```python
df_dates
```

```python
df_dates.reset_index(inplace=True,drop=True)
df_dates.to_csv("CarRegistrationDataReady.csv",index=False)
```

### Inflation data

```python
df_inflation = pd.read_excel("ExternalData\Inflation.xls")
```

```python
df_inflation
```

```python
df_inflation["Date"] = df_inflation["Year"].astype(str) + "-" + df_inflation['Month'].astype(str).str[0:2] + "-01"
df_inflation.drop(labels=["Month","Year"],axis=1,inplace=True)
```

```python
# Convert to datetime object and add new column indicating when the data is available (t+1 months)
df_inflation["Date"]= pd.to_datetime(df_inflation['Date'])
df_inflation["DataReleaseDate"] = df_inflation["Date"] + pd.to_timedelta(32,unit='d')
```

```python
# Set the data release date to nearest sunday
df_inflation["DataReleaseDate"] = df_inflation["DataReleaseDate"] - pd.to_timedelta(df_inflation["DataReleaseDate"].dt.dayofweek, unit='d') +  pd.to_timedelta(6, unit='d')
```

```python
df_inflation
```

- Put these into a weekly df

```python
df_dates = pd.date_range('2014-01-05', '2019-12-30', freq='1W').to_frame(name='Date').reset_index(drop=True)
df_dates.columns = ["Date"]
df_dates.sort_values(by="Date",inplace=True)
df_dates.reset_index(inplace=True,drop=True)
```

```python
df_dates["MoMInflation%"] = 0

for i in range(len(df_dates)):
    for k in range(len(df_inflation)-1):
        if df_dates.loc[i,"Date"] > df_inflation.loc[k,"DataReleaseDate"] and df_dates.loc[i,"Date"] <= df_inflation.loc[k+1,"DataReleaseDate"]:
            df_dates.loc[i,"MoMInflation%"] = df_inflation.loc[k,"MoMInflation%"]
```

```python
# Add one day so it matches with the training/test data
df_dates["Date"] = df_dates["Date"] + pd.to_timedelta(1,unit='d')
```

```python
df_dates.head()
```

```python
df_dates.reset_index(inplace=True,drop=True)
df_dates.to_csv("InflationDataReady.csv",index=False)
```

# GDP Growth Data

```python
df_gdp = pd.read_excel("ExternalData\GDPGrowth.xlsx")
df_gdp.columns = ["DataReleaseDate", "GDPGrowth%"]
```

```python
df_gdp.head()
```

```python
df_gdp["DataReleaseDate"] = pd.to_datetime(df_gdp['DataReleaseDate'])
```

```python
df_gdp.sort_values(by="DataReleaseDate",inplace=True)
df_gdp.reset_index(inplace=True,drop=True)
df_gdp.head(-1)
```

- Put these into a weekly df

```python
df_dates = pd.date_range('2014-01-05', '2019-12-30', freq='1W').to_frame(name='Date').reset_index(drop=True)
df_dates.columns = ["Date"]
df_dates.sort_values(by="Date",inplace=True)
df_dates.reset_index(inplace=True,drop=True)
```

```python
df_dates["GDPGrowth%"] = 0

for i in range(len(df_dates)):
    for k in range(len(df_gdp)-1):
        if df_dates.loc[i,"Date"] > df_gdp.loc[k,"DataReleaseDate"] and df_dates.loc[i,"Date"] <= df_gdp.loc[k+1,"DataReleaseDate"]:
            df_dates.loc[i,"GDPGrowth%"] = df_gdp.loc[k,"GDPGrowth%"]

    # set last 4 entries manually
    if i >= 309:
        df_dates.loc[i,"GDPGrowth%"] = df_gdp.loc[len(df_gdp)-1,"GDPGrowth%"]
```

```python
# Add one day so it matches with the training/test data
df_dates["Date"] = df_dates["Date"] + pd.to_timedelta(1,unit='d')
df_dates
```

```python
df_dates.reset_index(inplace=True,drop=True)
df_dates.to_csv("GDPDataReady.csv",index=False)
```
